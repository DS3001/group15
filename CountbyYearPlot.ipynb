```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('/cleaned_data.csv', low_memory = True)
# group by count plot

plt.figure(figsize=(8, 6))  # Set the figure size (optional)
sns.countplot(x='year', data=df)

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Count Plot Grouped by Year')

# Display the plot
plt.show()
```
