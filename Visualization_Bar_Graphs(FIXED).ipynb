```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/cleaned_data.csv', low_memory = True)
```

```python
import seaborn as sns
var = 'class'
sns.histplot(df[var])
print(df[var].describe)
plt.xlabel("Economic Class")
plt.ylabel("Count of respondents")
```

```python
svar = 'finalter'
sns.histplot(df[svar])
print(df[svar].describe)
plt.xlabel("Trend in Financial Status")
plt.ylabel("Count of Respondents")
```

```python
variables = 'satfin'
sns.histplot(df[variables])
print(df[variables].describe)
plt.xlabel("Satisfaction with Financial Siuation")
plt.ylabel("Count of Respondents")
```

```python

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.countplot(x="class", hue="sex", data=df, palette="Set2")
plt.title("Financial Situation by Gender")
plt.xlabel("satfin")
plt.ylabel("Count")
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.legend(title="sex")
plt.show()
```
