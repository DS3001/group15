```python
# making a kernel density plot of stroke grouped by age 
import seaborn as sns
import matplotlib.pyplot as plt
y = 'stroke'
gender = 'gender'
sns.kdeplot(data=df, x=y, hue = gender, palette = 'rocket') 
plt.title("KDE Plot of Stroke by Gender")
```

```python
# This is a kernel density plot that illustrates the density of the individuals in the data set by their gender and whether they experienced a stroke or not
# There is a greater density of females who have not experienced a stroke compared to their male and non binary counterparts
#There is also a greater density of women who experienced a stroke
# going to make a kernel density plot of stroke grouped by smoking status
```

```python
smoking = 'smoking_status'
sns.kdeplot(data=df,x=y,hue=smoking, palette = 'rocket')
plt.title('KDE Plot of Stroke by Smoking Status')
```

```python
# This is a kernel density plot of stroke and smoking status
# greatest density of individuals who never smoked that have experienced a stroke
# greatest density of individuals who's smoking status is unknown have not experienced a stroke
```

```python
# variables i am working with are : gender(categorical) age(numerical) hypertension(binary) heart_disease(binary) and smoking_status(categorical)
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.countplot(x="stroke", hue="hypertension", data=df, palette="rocket")
plt.title("Bar Graph of Stroke by Hypertension")
plt.xlabel("Experience of a Stroke")
plt.ylabel("Count")
plt.show()
```

```python
# bar graph depicting the count of people who experience (or not) a stroke on wether they have hypertension or not
# greater number of people who do not have hypertension have not experienced a stroke
# greater number of peope who do not have hypertension have experienced a stroke
```
