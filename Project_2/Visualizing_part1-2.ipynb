# Visualizing

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Data/training_cleaned.csv', low_memory = True)
```

```python
df.columns.tolist()
```

```python
## Bar graph for Stroke Likelihood by Gender

colors = sns.color_palette("Set1")
sns.set(style="whitegrid")
plt.figure(figsize=(15, 10))
sns.countplot(x="stroke", hue="gender", data=df, palette=colors)
plt.title("Stroke Likelihood by Gender",fontsize=20)
plt.xlabel("Stroke",fontsize=16)
plt.ylabel("Count",fontsize=16)
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.legend(title="gender")
plt.show()
```

^ Comparison of the Male/Female Non stroke and stroke experiencers in the data. 

```python
# Set the figure size
plt.figure(figsize=(15, 10))

## Graphing the KDE plot by Class

sns.kdeplot(data=df, x='age', hue = 'gender') # By class -- all
plt.title("KDE Plot by Age",fontsize=20)
plt.xlabel("Age")


# Display the plot (for interactive environments)

plt.show()
```

^ Density plot of the age of participants male and female. 

```python
# Set the figure size
plt.figure(figsize=(15, 10))

## Graphing the KDE plot by Class

sns.kdeplot(data=df, x='avg_glucose_level', hue = 'gender') # By class -- all
plt.title("KDE Plot by Glucose Level",fontsize=20)
plt.xlabel('Glucose Level (mg/dL)')


# Display the plot (for interactive environments)

plt.show()
```

^ Density plot of the glucose level across the data set between males and females

```python
# Set the figure size
plt.figure(figsize=(15, 10))

## Graphing the KDE plot by Class

sns.kdeplot(data=df, x='bmi', hue = 'gender') # By class -- all
plt.title("KDE Plot by BMI",fontsize=20)
plt.xlabel('BMI')


# Display the plot (for interactive environments)

plt.show()
```

^ density plot of BMI among males and females in the dataset

```python

# Sample data
labels = ['Male','Female']
sizes = [1611, 2317]

plt.figure(figsize=(15, 10))

colors = ['skyblue', 'lightcoral']
# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90,colors=colors)

# Set plot title
plt.title('Percent of Males and Females in the Dataset',fontsize=18)

# Show the plot
plt.show()

```

```python
## Bar graph for Marriage Status and stroke

sns.set(style="whitegrid")
plt.figure(figsize=(15, 10))
sns.countplot(x="stroke", hue="ever_married", data=df, palette="Set1")
plt.title("Stroke Occurence by Marriage Status",fontsize=20)
plt.xlabel("Stroke Occurence",fontsize=16)
plt.ylabel("Count",fontsize=16)
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.legend(title="Ever Married")
plt.show()
```

^ apparently people who have been married are more likely to get a stroke

```python
## Bar graph for Work Type and stroke 

sns.set(style="whitegrid")
plt.figure(figsize=(15, 10))
sns.countplot(x="stroke", hue="work_type", data=df, palette="Blues")
plt.title("Stroke Occurence by Work Type",fontsize=20)
plt.xlabel("Stroke Occurence",fontsize=16)
plt.ylabel("Count",fontsize=16)
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.legend(title="Work Type")
plt.show()
```

^ bar graph of all the work types

```python
## Bar graph for Residence Type and stroke 

sns.set(style="whitegrid")
plt.figure(figsize=(15, 10))
sns.countplot(x="stroke", hue="Residence_type", data=df, palette="B")
plt.title("Stroke Occurence by Residence Type",fontsize=20)
plt.xlabel("Stroke Occurence",fontsize=16)
plt.ylabel("Count",fontsize=16)
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.legend(title="Residence Type")
plt.show()
```

^ stroke likelihood is about the same for rural and urban areas

```python
# Scatterplot using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x='avg_glucose_level', y='age', data=df, hue='stroke', palette='Set1', marker='o',alpha =.6)

# Set labels and title
plt.xlabel('Blood Glucose Levels')
plt.ylabel('Age')
plt.title('Stroke Likelihood by Blood Glucose Levels and Age',fontsize=18)

# Show the plot
plt.show()

# chatgpt helped make this graph
```

Shows that there's not really a trend for people with specifically high blood glucose having a stroke

```python
# Scatterplot using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x='bmi', y='age', data=df, hue='stroke', palette='RdPu', marker='o')

# Set labels and title
plt.xlabel('BMI')
plt.ylabel('Age')
plt.title('Stroke Likelihood by BMI and Age',fontsize=18)

# Show the plot
plt.show()

```

Shows that age might have more of an impact on stroke likelihood than BMI

```python
# Scatterplot using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x='bmi', y='age', data=df, hue='gender', style='stroke', palette='Set1',alpha =.7)

# Set labels and title
plt.xlabel('BMI')
plt.ylabel('Age')
plt.title('Stroke Likelihood by BMI, Age, and Gender',fontsize=18)

# Show the plot
plt.show()
```

The xs are stroke occurrences. male is red. female blue. circle is no stroke
