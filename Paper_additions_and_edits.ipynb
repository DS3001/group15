## Data: 

Some background on the survey is that each year, the data is collected by the National Opinion Research Center, and is a panel that records data from the same 3000 people over time. One thing to note about the number of people surveyed is that 3000 is the intended number of respondents, but this number fluctuates anywhere from 1250 to nearly 4000 in our data. The questions within the survey aim to identify the attitudes and behaviors of participants along with demographic data. Demographic data takes up around 30% while the attitudes and behaviors portion takes up the remainder of the survey. There are “core questions” which are asked each year, and there are “topical modules” which ask questions relevant to the time period, for example, attitudes on abortion in 1977. 

An important point to note about these data is that the responses are extremely subjective. Because there is no way to determine whether respondents are telling the truth, there is a possibility that not all of the data adequately represents the population. Specifically in the instances of class reporting, satisfaction with financial situation, and trend in financial situation, participants may be compelled to self-identify as working class rather than middle class, having “better” rather than “stayed same” finances, or “more or less satisfied” rather than “not satisfied at all” with their finances either because they simply do not want to admit that they are unsatisfied or because they do not want to reveal their actual circumstances. Either way, this adds an interesting element of nuance to the evaluation of these data and its explanatory properties that must be taken into account when drawing conclusions about the general attitudes of the population. 

It is also important to note that each of these responses is self-reported, and the verbiage of each response possibility can influence the choice of the participant. For example, it may just be easier to select “more or less satisfied” than “pretty well satisfied” because “pretty well satisfied” is not a commonly used way to describe current job satisfaction.
