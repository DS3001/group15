```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/cleaned_data.csv', low_memory = True)
```

```python
df.columns.tolist()
```

```python
var1 = 'sex' #need numeric -- gender
var2 = 'year'
cat1 = 'class'
cat2 = 'finalter' # finalter
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.countplot(x="satfin", hue="sex", data=df, palette="Set2")
plt.title("Financial Situation by Gender")
plt.xlabel("satfin")
plt.ylabel("Count")
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.legend(title="sex")
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.countplot(x="year", hue="satfin", data=df, palette="Set2")
plt.title("Financial Situation by Year")
plt.xlabel("satfin")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Satisfaction")
plt.show()
```

```python
male_df = df[df['sex'] == 'male']
female_df = df[df['sex'] == 'female']

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
```

```python
#male graph
sns.countplot(x="year", hue="satfin", data=male_df, palette="Set2")
plt.title("Financial Situation by Year (Males)")
plt.xlabel("Year")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Satisfaction")
```

```python
# female graph
sns.countplot(x="year", hue="satfin", data=female_df, palette="Set2")
plt.title("Financial Situation by Year (Females)")
plt.xlabel("Year")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Satisfaction")
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.countplot(x="year", hue="class", data=male_df, palette="Set2")
plt.title("Subjective class placement by Year (Male)")
plt.xlabel("class")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Class")
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.countplot(x="year", hue="class", data=female_df, palette="Set2")
plt.title("Subjective class placement by Year (Female)")
plt.xlabel("class")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Class")
plt.show()
```
